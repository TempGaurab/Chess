```python
def to_5bit_binary(value, offset):
    return format(value + offset, '06b')

# Define offsets for letters and digits
letter_offset = 0
digit_offset = 26

# Generate mappings for a-z (letters)
letter_mapping = {chr(i): to_5bit_binary(i - ord('a'), letter_offset) for i in range(ord('a'), ord('z') + 1)}

# Generate mappings for 0-9 (digits) with offset
digit_mapping = {str(i): to_5bit_binary(i, digit_offset) for i in range(10)}

# Combine both mappings
binary_mapping = {**letter_mapping, **digit_mapping}

# Print the complete binary mapping
print(binary_mapping)
```

```python
def message_to_binary(message, binary_mapping):
    # Convert each character in the message to its binary representation
    binary_message = ''.join(binary_mapping[char] for char in message)
    return binary_message
```

```python
message = "hello"
binary_message = message_to_binary(message, binary_mapping)
binary_message
```

```python
chunks = [binary_message[i:i+6] for i in range(0, len(binary_message), 6)]
chunks
```

```python
reverse_mapping = {v: k for k, v in binary_mapping.items()}
reverse_mapping
```

```python
decoded_message = ''.join([reverse_mapping[chunk] for chunk in chunks])
decoded_message
```

```python
def handle_move(board, move):
    try:
        # Try to make the move
        chess_move = chess.Move.from_uci(move)
        
        if chess_move in board.legal_moves:
            # Make the move
            board.push(chess_move)
            print(f"Move made: {move}")
        else:
            print("Invalid move!")
            return

    except Exception as e:
        print(f"Error: {e}")
        return

    # Check if the move is the first or last valid move
    legal_moves_list = list(board.legal_moves)
    if len(legal_moves_list) > 0:
        if move == legal_moves_list[0].uci():
            print("First valid move: True")
        elif move == legal_moves_list[-1].uci():
            print("Last valid move: False")

    # Check for stalemate
    if board.is_stalemate():
        print("Stalemate detected. Resetting the board.")
        board.reset()
    
    # Check for game over
    elif board.is_game_over():
        print("Game over. Resetting the board.")
        board.reset()
```

```python
# Example usage
board = chess.Board()
board
```

```python
handle_move(board, "e2e4")  # Example of making a move
board
```
